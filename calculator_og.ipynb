```python
##                             MY CALCULATOR                                ##
def calculatro():
    print("\== PYTHON SCIENTIFIC CALCULATOR ==/")
    print('''
    1.Addition              13.Natural Log
    4.Division              14.Base Log
    5.Modulus               15.Exponential
    6.Power                 16.Exit
    7.Square Root           17.View Last 5 Calculations :
    2.Substraction          18.View Session History :
    3.Mulltiplication       19.delete and clear history file :
    8.Cube Root
    9.Factorial
    10.Sin()
    11.Cos()
    12.Tan()
    ''')

    import os
    import math
    while True:
        try:
            d=int(input("Enter choice"))
            if d<20:
                if d==16:
                    print("exited..!")
                    break
                elif d==1 or d==2 or d==3 or d==4 or d==5 or d==6:
                    a=int(input("Eter 1st no:"))
                    c=int(input("Enter 2nd no:"))
                    if d==1:
                        e=a+c
                        with open("calculatro.txt","a")as y:
                            y.write(f'{a}+{c}={e}\n')
                        print(a,"+",c,"=",a+c)
                    elif d==2:
                        f=a-c
                        with open("calculatro.txt","a")as y:
                            y.write(f'{a}-{c}={f}\n')
                        print(a,"-",c,"=",a-c)
                    elif d==3:
                        g=a*c
                        with open("calculatro.txt","a")as y:
                            y.write(f'{a}*{c}={g}\n')
                        print(a,"*",c,"=",a*c)
                    elif d==4:
                        h=a/c
                        with open("calculatro.txt","a")as y:
                            y.write(f'{a}/{c}={h}\n')
                        print(a,"/",c,"=",a/c)
                    elif d==5:
                        i=a%c
                        with open("calculatro.txt","a")as y:
                            y.write(f'{a}%{c}={i}\n')
                        print(a,"%",c,"=",a%c)
                    elif d==6:
                        j=a**c
                        with open("calculatro.txt","a")as y:
                            y.write(f'{a}**{c}={j}\n')
                        print(a,"**",c,"=",math.pow(a,c))
                else:
                    if d==7 or d==8 or d==9 or d==10 or d==11 or d==12 or d==13 or d==14 or d==15:
                        b=int(input("enter your number"))
                        e=math.radians(b)
                        if d==7:
                            k=math.sqrt(b)
                            with open("calculatro.txt","a")as y:
                                y.write(f'sqrt of :{b}= {k}\n')
                            print("sqrt of :",b,"=",math.sqrt(b))
                        elif d==8:
                            with open("calculatro.txt","a")as y:
                                y.write(f'cbrt of :{b}= {math.cbrt(b)}\n')
                            print("cbrt :",b,"=",math.cbrt(b))
                        elif d==9:
                            with open("calculatro.txt","a")as y:
                                y.write(f'factorial of :{b}= {math.factorial(b)}\n')
                            print("factorial :",b,"=",math.factorial(b))
                        elif d==10:
                            with open("calculatro.txt","a")as y:
                                y.write(f'sin of :{b}= {math.sin(e)}\n')
                            print("sin :",b,"=",math.sin(e))
                        elif d==11:
                            with open("calculatro.txt","a")as y:
                                y.write(f'cos of :{b}= {math.cos(e)}\n')
                            print("cos :",b,"=",math.cos(e))
                        elif d==12:
                            with open("calculatro.txt","a")as y:
                                y.write(f'tan of :{b}= {math.tan(e)}\n')
                            print("tan :",b,"=",math.tan(e))
                        elif d==13:
                            with open("calculatro.txt","a")as y:
                                y.write(f'natural log of :{b}= {math.log(b)}\n')
                            print("natural log :",b,"=",math.log(b))
                        elif d==14:
                            with open("calculatro.txt","a")as y:
                                y.write(f'log10 of :{b}= {math.log10(b)}\n')
                            print("log10 :",b,"=",math.log10(b))
                        elif d==15:
                            with open("calculatro.txt","a")as y:
                                y.write(f'exponential :{b}= {math.exp(b)}\n')
                            print("exponential :",b,"=",math.exp(b))
                if d==17:
                    with open("calculatro.txt","r") as z:
                        k=z.readlines()
                        print("Last 5 History : ")
                        print(k[-5:])
                elif d==18:
                    with open("calculatro.txt","r") as z:
                        print("History :")
                        print(z.read())
                elif d==19:
                    os.remove("calculatro.txt")
                    print("history file deleted")
            else:
                print("Enter correct choice 1 to 19")
        except Exception as ww:
            print(ww)
            print("Try Again...!")
        else:
            print("calculation completed..")
        finally:
            print("Thanks")
calculatro()
```
